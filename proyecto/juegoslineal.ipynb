```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import statsmodels.api as sm
from sklearn.preprocessing import PolynomialFeatures
```

## Simple linear regression 

```python
plt.figure(figsize=(16, 8))
plt.scatter(
    data['Global_Sales'],
    data['NA_Sales'],
    c='black'
)
plt.xlabel("global sales")
plt.ylabel("Ventas ($)")
plt.show()
```

```python
X = data['Global_Sales'].values.reshape(-1,1)
y = data['NA_Sales'].values.reshape(-1,1)

reg = LinearRegression()
reg.fit(X, y)
```

```python
print(reg.coef_[0][0])
print(reg.intercept_[0])

print("The linear model is: Y = {:.5} + {:.5}X".format(reg.intercept_[0], reg.coef_[0][0]))
```

```python
predictions = reg.predict(X)

plt.figure(figsize=(16, 8))
plt.scatter(
    data['Global_Sales'],
    data['NA_Sales'],
    c='black'
)
plt.plot(
    data['Global_Sales'],
    predictions,
    c='blue',
    linewidth=2
)
plt.xlabel("Money spent on TV ads ($)")
plt.ylabel("Sales ($)")
plt.show()
```

```python
X = data['Global_Sales']
y = data['NA_Sales']

X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```

```python
X = np.column_stack((data['EU_Sales'], data['Global_Sales'], data['Critic_Score']))
y = data['NA_Sales']

X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```

### Interaction terms

```python
# Create interaction term (not polynomial features)
interaction = PolynomialFeatures(degree=3, include_bias=False, interaction_only=True)
X_inter = interaction.fit_transform(X)
#X_test_inter = interaction.fit_transform(X_test)
```

```python
X2 = sm.add_constant(X_inter)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```
